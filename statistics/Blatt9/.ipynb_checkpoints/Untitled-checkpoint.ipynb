## Aufgabe 24

```r
count = c(355, 211, 197, 166, 142, 106, 104, 60, 56, 38, 36, 32, 21, 19, 15)
interval = 1:15

rays = data.frame(counts=count, intervall=interval)
plot(rays)
rays$counts = log(rays$counts)
```

```r
rays.lm = lm(rays)
summary(rays.lm)
plot(rays$counts ~ rays$intervall)
abline(rays.lm, col="red")
```

```r
install.packages("ellipse")
library(ellipse)
```

```r
confidence.intervalls = confint(rays.lm)
confidence.intervalls
confidence.ellipse = el=ellipse(rays.lm, level=0.95, which=c(1,2))
plot(confidence.ellipse, pch=20, col="blue")
points(rays.lm$coefficients[1], rays.lm$coefficients[2], pch=16,lwd=3,col="red")
rect(confidence.intervalls[1,1], confidence.intervalls[2,1],
     confidence.intervalls[1,2], confidence.intervalls[2,2],
    border="blue")
```

```r
start.state = data.frame(intervall=0)
start.count = predict(rays.lm, start.state, interval="prediction")

new.count = c(exp(start.count)[1], count)
new.intervalls = 0:15

predicted = data.frame(counts=new.count, intervall=new.intervalls)
plot(predicted)

rate.prediction = rays.lm$coefficients[2]

#rate.points = new.count + new.count * rate.prediction
#lines(rate.points, new.intervalls, col="blue")
rate.function = sapply(0:15, function(x) {new.count[1] * (1 + rate.prediction)^x})
lines(rate.function, new.intervalls, col="red")

nine.state = data.frame(intervall=1.5)
nine.count = predict(rays.lm, nine.state, interval="prediction")
nine.count
points(exp(nine.count)[1], 1.5, col="green", bg="green")
lines(exp(nine.count)[2:3], c(1.5, 1.5) , col="green")
```

```r
rays$offset = 0
rays
plot(rays$intervall, rays$counts)
abline(lm(rays$counts ~ I(rays$intervall - log(400)) - 1, offset=rays$offset))
```

```r
library(MASS)
data(cement)
cement
summary(cement)
pairs(cement)
```

```r
fit.cement = lm(y ~ x1 + x2 + x3 + x4, data=cement)
summary(fit.cement)
# Std error of intercept very high
```

```r
cor(cement)
```

```r

attach(cement)
t.value = function(form) {
    a = summary(lm(formula(form)))
    coef(a)[-1, "t value"]
}

test = c("x1", "x2", "x3", "x4")
l1 = c("y ~ x1", "y ~ x2", "y ~ x3", "y ~ x4")
mins = sapply(l1, t.value)
mins
b = l1[which.min(mins)]
l2 = sapply(test, function(x, best) { paste(best,x, sep="+")}, best=b)
l2
mins = sapply(l2, t.value)
b = l2[which.min(mins)]


detach(cement)
```
